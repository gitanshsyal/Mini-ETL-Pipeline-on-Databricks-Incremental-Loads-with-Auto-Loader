```python
from pyspark.sql.functions import *
##bronze_df=spark.read.format('delta').load('abfss://catalog@sourcestorageprj.dfs.core.windows.net/sales/bronze')
bronze_df=spark.readStream.format('delta').load('abfss://catalog@sourcestorageprj.dfs.core.windows.net/sales/bronze/data')
display(bronze_df)
```

```python
dbServer = 'db-server1'
dbPort = '1433'
dbName = 'db-name1'
databricksScope = 'project-scope'
dbUser = dbutils.secrets.get(scope = databricksScope, key='username')
dbPassword = dbutils.secrets.get(scope = databricksScope, key='password')
connectionUrl ='jdbc:sqlserver://{}.database.windows.net:{};database={};user={};'.format(dbServer,dbPort, dbName, dbUser)
connectionProperties = {
'password': dbPassword,
'driver':'com.microsoft.sqlserver.jdbc.SQLServerDriver'
}
valid_customers_df = spark.read.jdbc(url = connectionUrl, table =
'dbo.valid_customers',
properties = connectionProperties)
valid_products_df = spark.read.jdbc(url = connectionUrl, table =
'dbo.valid_products',
properties = connectionProperties)
customers_list=valid_customers_df.select('customer_id').collect()
valid_customers_list=[row['customer_id'] for row in customers_list]
product_list=valid_products_df.select('product_id').collect()
valid_product_list=[row['product_id'] for row in product_list]


```

```python
valid_df=bronze_df.withColumn('category',when((col('Quantity')>0) & (col('Price')>0) &(col('CustomerID').isin(valid_customers_list)) &(col('ProductID').isin(valid_product_list)),'valid')\
                              .otherwise('bad'))
valid_df=valid_df.withColumn('rejection_reason',when(col('quantity')<0,'Invalid_quanity')\
    .otherwise(when(col('price')<0,'Invalid_price')\
        .otherwise(when(~col('CustomerID').isin(valid_customers_list),'Invalid_customer')\
            .otherwise(when(~col('ProductID').isin(valid_product_list),'Invalid_product')\
                .otherwise('NA')))))
logs_df=valid_df.filter(col('category')=='bad').groupBy('rejection_reason').agg(count('*').alias('count_type'))
logs_df.writeStream \
    .format("delta") \
    .option("checkpointLocation", "abfss://catalog@sourcestorageprj.dfs.core.windows.net/sales/silver/bad/data_report/_checkpoint") \
    .outputMode("complete") \
    .trigger(availableNow=True) \
    .start('abfss://catalog@sourcestorageprj.dfs.core.windows.net/sales/silver/bad/data_report')
display(logs_df)
```
