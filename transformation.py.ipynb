```python
silver_df = (
    spark.readStream
         .format("delta")
         .load("abfss://catalog@sourcestorageprj.dfs.core.windows.net/sales/silver/data")
         .where("category = 'valid'")
)
```

```python
from pyspark.sql.functions import *
modified_df=silver_df.withColumn('total_amount',col('quantity')*col('price'))
display(modified_df)
```

```python
dbServer = 'db-server1'
dbPort = '1433'
dbName = 'db-name1'
databricksScope = 'project-scope'
dbUser = dbutils.secrets.get(scope = databricksScope, key='username')
dbPassword = dbutils.secrets.get(scope = databricksScope, key='password')
connectionUrl ='jdbc:sqlserver://{}.database.windows.net:{};database={};user={};'.format(dbServer,dbPort, dbName, dbUser)
connectionProperties = {
'password': dbPassword,
'driver':'com.microsoft.sqlserver.jdbc.SQLServerDriver'
}
valid_customers_df = spark.read.jdbc(url = connectionUrl, table =
'dbo.valid_customers',
properties = connectionProperties)
valid_products_df = spark.read.jdbc(url = connectionUrl, table =
'dbo.valid_products',
properties = connectionProperties)
display(valid_products_df)


```

```python
modified_product_df=modified_df.join(valid_products_df,modified_df.ProductID==valid_products_df.product_id,how='inner').drop('valid_products_df.product_id','category')
modified_customer_df=modified_product_df.join(valid_customers_df,modified_product_df.CustomerID==valid_customers_df.customer_id,how='inner').drop('valid_customers_df.customer_id')
```

```python
modified_customer_df.writeStream.format('delta').option('checkpointLocation','abfss://catalog@sourcestorageprj.dfs.core.windows.net/sales/silver/target/checkpoint').outputMode('append').option('mergeSchema','true').trigger(availableNow=True).start('abfss://catalog@sourcestorageprj.dfs.core.windows.net/sales/silver/target/data')
```
