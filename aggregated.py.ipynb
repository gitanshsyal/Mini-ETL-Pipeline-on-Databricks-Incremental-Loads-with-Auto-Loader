```python
source_df=spark.read.format('delta').load('abfss://catalog@sourcestorageprj.dfs.core.windows.net/sales/silver/target/data')
display(source_df)
```

```python
from pyspark.sql.functions import *
top_three_customers_df=source_df.groupBy('CustomerID','customer_name').agg(sum('total_amount').alias('total_amount_spend')).orderBy(col('total_amount_spend'),ascending=False).limit(3)
total_sales_category_state_df=source_df.groupBy('product_category','State').agg(sum('total_amount').alias('total_sales')).orderBy(col('total_sales'),ascending=False)
daily_revenue_df=source_df.groupBy('Date').agg(sum('total_amount').alias('daily_revenue')).orderBy('Date')
display(daily_revenue_df)
```

```python
top_three_customers_df.write.format('csv').mode('overwrite').option('header',True).save('abfss://catalog@sourcestorageprj.dfs.core.windows.net/sales/gold/top_three_customers')
total_sales_category_state_df.write.format('csv').mode('overwrite').option('header',True).save('abfss://catalog@sourcestorageprj.dfs.core.windows.net/sales/gold/revenue_by_category_state')
daily_revenue_df.write.format('csv').mode('overwrite').option('header',True).save('abfss://catalog@sourcestorageprj.dfs.core.windows.net/sales/gold/daily_revenue')
```
